# Data Cleaning with Python


```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("shopify_raw.csv")
data_raw = data.copy() 
```

```python
data_raw['order_date'] = pd.to_datetime(data_raw['order_date'])
data_raw.info()
# data: 60000 rows, 17 columns
```

```python
# check data
def check_data(data):
    quality_dict = {
        'missingvalue': data.isnull().sum().to_dict(),
        'duplication': data.duplicated().sum(),
        'data info': data.info() 
    }
    return quality_dict
```

```python
response = check_data(data_raw)
display(response)
```

#### RFM analysis

```python
# tong doanh thu cua tung khach hang
dt = data_raw.copy()
dt['total_revenue'] = data_raw.groupby('customer_id')['revenue'].transform('sum')
```

```python
Q1 = dt['total_revenue'].quantile(0.25)
Q3 = dt['total_revenue'].quantile(0.75)

IQR = Q3-Q1

print(IQR)
print(Q1)
print(Q3)
```
